There is still hope for a good model though because the scatter plot only shows the first two principal components, which according to this curve, only explains a very small percent of total variance.

## split data into test and training

```python
df.shape
```

```python
mapping = {'right':1, 'left':0}
Y = np.array(df['lean'].apply(lambda x: mapping[x])).T
Y.shape
```

```python
X = document_vectors
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
print('X_train: ',X_train.shape)
print('X_test: ',X_test.shape)
print('Y_train: ',Y_train.shape)
print('Y_test: ',Y_test.shape)
```

## Train Basic models

```python


# Preprocessing (Label Encoder)
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Basic Models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Ensemble Models
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

# Validation
from sklearn.model_selection import train_test_split, cross_val_score, KFold

# Model Parameter Tuning
from sklearn.model_selection import GridSearchCV

# ML Scoring Metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
models = [
    ( 'Logistic Regression' , LogisticRegression()),
    ( 'K-Nearest Neighbors' , KNeighborsClassifier()),
    ( 'Naive Bayes' , GaussianNB()),
    ( 'Support Vector Machine' , SVC())
]

results = []
names = []

for name, model in models:
    # training with 10-fold cross validation
    Kfold = KFold(n_splits=10, random_state=3)
    cv_results = cross_val_score(model, X_train, Y_train, cv=Kfold, scoring= 'accuracy')
    
    # save results
    results.append(cv_results)
    names.append(name)

    # print out results
    print(f'{name}: \t {cv_results.mean()} +/- {cv_results.std()}')
```

## training ensemble models

```python
ensemble_models = [
    ( 'AdaBoost' , AdaBoostClassifier()),
    ( 'Bagging' , BaggingClassifier()),
    ( 'Gradient Boosted' ,  GradientBoostingClassifier()),
    ( 'Random Forest' , RandomForestClassifier())
]

for name, model in ensemble_models:
    # training with 10-fold cross validation
    Kfold = KFold(n_splits=10, random_state=3)
    cv_results = cross_val_score(model, X_train, Y_train, cv=Kfold, scoring= 'accuracy')
    
    # save results
    results.append(cv_results)
    names.append(name)

    # print out results
    print(f'{name}: \t {cv_results.mean()} +/- {cv_results.std()}')
```

## xgboost

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
name='XGBoost'

# training with 10-fold cross validation
Kfold = KFold(n_splits=10, random_state=3)
cv_results = cross_val_score(model, X_train, Y_train, cv=Kfold, scoring= 'accuracy')

# save results
results.append(cv_results)
names.append(name)

# print out results
print(f'{name}: \t {cv_results.mean()} +/- {cv_results.std()}')
```

## auc roc curve

```python
from sklearn.metrics import roc_curve, auc



models_for_auc_curves={}
models_for_auc_curves['Gradient Boosted']= GradientBoostingClassifier()
models_for_auc_curves['XGBoost']=XGBClassifier()
models_for_auc_curves['Support Vector Machine'] = SVC(probability=True)
models_for_auc_curves['RandomForest']=RandomForestClassifier()

trained_models = {}

for classifier in models_for_auc_curves:

    trained_classifier=models_for_auc_curves[classifier].fit(X_train, Y_train)

    trained_models[classifier] = trained_classifier

    Y_score = trained_classifier.predict_proba(X_test)[:, 1]

    fpr, tpr, _ = roc_curve(Y_test, Y_score)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, lw=2, label=f'{classifier} (AUC-ROC = {roc_auc:.3f})')
    print(f'Done with {classifier}: {roc_auc}')


plt.tick_params()
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc="lower right")
plt.gcf().set_size_inches(10, 10)
plt.savefig('../AUC-ROC_Curve.png', dpi=1200)
plt.show()
```

```python
# ML Scoring Metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import seaborn as sns


Y_pred = trained_models['XGBoost'].predict(X_test)

# Confusion Matrix
x_axis_labels = ['Left-lean', 'Right-lean'] # labels for x-axis
y_axis_labels = ['Left-lean', 'Right-lean'] # labels for y-axis
plt.figure(figsize=(8,8))
sns.heatmap(confusion_matrix(Y_test, Y_pred), annot=True, fmt='g',xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.title("Confusion Matrix",fontweight="bold")
plt.ylabel("True Ideological Lean")
plt.xlabel("Predicted Ideological Lean")
plt.savefig('../confusion-matrix.png', dpi=1200)
plt.show()
```

```python
# Classification Report

print(classification_report(Y_test,Y_pred))
```
